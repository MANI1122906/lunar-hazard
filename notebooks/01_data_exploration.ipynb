# 1. Data Exploration - Chandrayaan Lunar Satellite Imagery

This notebook explores the Chandrayaan satellite imagery dataset, including loading, visualization, and basic analysis of lunar surface features.

## Import Required Libraries

```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import seaborn as sns
import rasterio
import h5py
from pathlib import Path
from PIL import Image
import warnings
warnings.filterwarnings('ignore')

# Set style for visualizations
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
```

## Load and Explore Chandrayaan Data

```python
# Define data paths
raw_data_path = Path('../data/raw')
processed_data_path = Path('../data/processed')
masks_path = Path('../data/masks')

# List available data files
tif_files = list(raw_data_path.glob('*.tif'))
hdf5_files = list(raw_data_path.glob('*.hdf5'))

print(f"Found {len(tif_files)} .tif files")
print(f"Found {len(hdf5_files)} .hdf5 files")
print(f"\nTIF Files: {[f.name for f in tif_files]}")
print(f"HDF5 Files: {[f.name for f in hdf5_files]}")
```

```python
# Function to load and analyze .tif files
def analyze_tif_file(filepath):
    """Load and analyze a GeoTIFF file"""
    with rasterio.open(filepath) as src:
        data = src.read()
        profile = src.profile
        
    return data, profile

# Function to load and analyze .hdf5 files
def analyze_hdf5_file(filepath):
    """Load and analyze an HDF5 file"""
    with h5py.File(filepath, 'r') as f:
        print(f"HDF5 Keys: {list(f.keys())}")
        metadata = {}
        for key in f.keys():
            metadata[key] = f[key].shape
    return metadata

print("Functions defined for data loading")
```

## Data Statistics and Visualization

```python
# Example: Create synthetic lunar-like data for demonstration
# In practice, load real Chandrayaan data here

# Generate synthetic lunar surface image
height, width = 512, 512
synthetic_image = np.random.rand(height, width) * 255

# Add some terrain features (simulated boulders and slopes)
y, x = np.ogrid[0:height, 0:width]
for _ in range(5):
    cy, cx = np.random.randint(height), np.random.randint(width)
    radius = np.random.randint(20, 50)
    mask = (x - cx)**2 + (y - cy)**2 <= radius**2
    synthetic_image[mask] = np.random.rand() * 255

synthetic_image = synthetic_image.astype(np.uint8)

# Display image statistics
print(f"Image Shape: {synthetic_image.shape}")
print(f"Data Type: {synthetic_image.dtype}")
print(f"Min Value: {synthetic_image.min()}")
print(f"Max Value: {synthetic_image.max()}")
print(f"Mean Value: {synthetic_image.mean():.2f}")
print(f"Std Dev: {synthetic_image.std():.2f}")
```

```python
# Visualize the lunar surface image
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Display image
im = axes[0].imshow(synthetic_image, cmap='gray')
axes[0].set_title('Synthetic Lunar Surface', fontsize=14, fontweight='bold')
axes[0].set_xlabel('X Pixel')
axes[0].set_ylabel('Y Pixel')
plt.colorbar(im, ax=axes[0], label='Intensity')

# Display histogram
axes[1].hist(synthetic_image.ravel(), bins=50, color='skyblue', edgecolor='black')
axes[1].set_title('Intensity Distribution', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Pixel Intensity')
axes[1].set_ylabel('Frequency')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Visualization complete")
```
